```python
import pandas as pd
import numpy as np
```

```python
sample_data=pd.read_csv("sample_data.csv")
sample_data
```

```python
sample_data1=sample_data.drop("salary_more_then_100k", axis="columns")
sample_data1.head(10)
```

```python
smple_data= sample_data['salary_more_then_100k']
smple_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb_company=LabelEncoder()
lb_job=LabelEncoder()
lb_degree=LabelEncoder()
```

```python
sample_data1['company_n'] = lb_company.fit_transform(sample_data1['company'])
sample_data1['job_n'] = lb_job.fit_transform(sample_data1['job'])
sample_data1['degree_n'] = lb_degree.fit_transform(sample_data1['degree'])
```

```python
sample_data1
```

```python
test_data=sample_data1.drop(["company","job","degree"],axis="columns")
test_data
```

```python
smple_data
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(test_data,smple_data)
```

```python
model.score(test_data,smple_data)
```
